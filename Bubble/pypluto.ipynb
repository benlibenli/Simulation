```python
# Loading the relevant packages
import pyPLUTO

# Load data
Data = pyPLUTO.Load()

# Creating the image
Image = pyPLUTO.Image(figsize=[7, 6], nwin=2)

# x-y slice
rho2D = Data.rho[:,:,128]

# Plotting the data
Image.display(
    rho2D,
    x1=Data.x1r,
    x2=Data.x2r,
    xtitle="x",
    ytitle="y",
    title="Density at z=0",
    cpos="right",
    cmap="RdBu_r",
    clabel=r"$\rho$",
)

# Saving the image and showing the plot
Image.savefig("bubble.png")
pyPLUTO.show()
```
